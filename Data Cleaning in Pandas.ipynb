```python
df = pd.read_excel(r'D:\Universe\Data Analysis\Customer Call List.xlsx')
df
```

```python
# First Step of Data Cleaning: REMOVE DUPLICATES

df = df.drop_duplicates()
```

```python
# Second Step of Data Cleaning (Python): REMOVE UNUSEFUL COLUMNS

df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
# standardizing -- .str.strip -- takes away misc characters from outsides of value (default white space removal)
# specify values inside parameter 
# can't use list to make multiple arguments
# can specify direction using .lstrip or .rstrip

df['Last_Name'].str.strip('...')
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('...')

# when saving to df make sure to save only on column you are editing not the entire df
# otherwise you will overwrite entire df with just the adjdusted column
```

```python
# must do strip cleaning one by one
df['Last_Name'] = df['Last_Name'].str.strip('/')
df['Last_Name'] = df['Last_Name'].str.strip('_')
```

```python
df
```

```python
# Ultimate Strip
# typing all characters in one argument

df2 = pd.read_excel(r'D:\Universe\Data Analysis\Customer Call List.xlsx')
df2
```

```python
df2['Last_Name'] = df2['Last_Name'].str.strip('._/')
df2

# all characters/symbols typed into one expression, just once
```

```python
# CLEANING NUMBERS
# using .replace
# deleter of everything other than numbers and letters: [^a-zA-Z0-9], replace w/ blank ''

df2['Phone_Number'] = df2['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex = True)
df2

# replacing phone numbers with numbers only, no symbols
```

```python
# Lambda (review via Visually Explained)
# using .apply

#df2['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-')
df2['Phone_Number'] = df2['Phone_Number'].apply(lambda x: str(x))
df2

# error, floats/intergers not subscriptable
# changing ints to strings using a lambda function
```

```python
type(df2['Phone_Number'])
```

```python
df2['Phone_Number'] = df2['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df2
```

```python
# REMOVING N/A values

df2['Phone_Number'] = df2['Phone_Number'].str.replace('nan--', '')
df2['Phone_Number'] = df2['Phone_Number'].str.replace('Na--', '')
df2

# be VERY careful when running code, cause you may overwrite or double apply something with unintended consequences
```

```python
# STANDARDIZING ADDRESS
# using .str.split()

df2['Address'].str.split(',', n=2, expand=True)
```

```python
# you can define column names when you assign it to variable, and creating a list

df2[['Street_Address', 'State', 'Zipcode']] = df2['Address'].str.split(',', n=2, expand=True)
df2
```

```python
df2 = df2.drop(columns = 'Not_Useful_Column')
df2
```

```python
# STANDARDIZING Paying Customer and Do Not Contact columns

df2['Paying Customer'] = df2['Paying Customer'].str.replace('Yes', 'Y')
df2['Paying Customer'] = df2['Paying Customer'].str.replace('No', 'N')
df2
```

```python
df2['Do_Not_Contact'] = df2['Do_Not_Contact'].str.replace('Yes', 'Y')
df2['Do_Not_Contact'] = df2['Do_Not_Contact'].str.replace('No', 'N')
df2
```

```python
# STANDARDIZING all N/a or nulls in the table
# no need to use .str.replace when working through entire table, just .replace

df2 = df2.replace('N/a', '')
df2
```

```python
# NaN is an actual null valure and not a string to be replace, can not use .replace
# use .fillna instead

df2 = df2.fillna('')
df2
```

```python
# DROPPING unnecessary columns
# in this case we used a for loop to loop through the df index and used loc to find where Do Not Contact = Yes)
# wild

for x in df2.index:
    if df2.loc[x, 'Do_Not_Contact'] == 'Y':
        df2.drop(x, inplace = True)

df2
```

```python
# removing rows with blank phone numbers

for x in df2.index:
    if df2.loc[x, 'Phone_Number'] == '':
        df2.drop(x, inplace = True)

df2

# to remove null values / blanks -- you can also use .dropna(subset = 'column name')
```
